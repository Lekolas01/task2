```python
import csv
import numpy as np

labels = ["age","workclass","education_num","occupation","race","sex","hours_per_week","country"]
indices = [0,1,4,6,8,9,12,13]

filename = "adult"
f_train_pos = open(filename+".f", "w")
f_train_neg = open(filename+".n", "w")

train_size = 2000
test_size = 1000

def _format(x):
    x = x.lower().replace("-","_").replace("&","_").strip()
    return x
    

with open('adult.data', 'r') as file:
    reader = csv.reader(file)
    for i in range(train_size):
        curr = np.array(next(reader))
        #print(labels)
        #print(first[indices])
        
        row = curr[indices]
        #print("x"+curr[-1]+"x")
        if any(word == ' ?' for word in row): continue

        if curr[-1] == " >50K":
            #print(curr[-1])
            f_train_pos.write(filename+"("+",".join(map(lambda x: _format(x), row))+").\n")
        else:
            f_train_neg.write(filename+"("+",".join(map(lambda x: _format(x), row))+").\n")
            
f_train_pos.close()
f_train_neg.close()



f_test_pos = open(filename+"_test.f", "w")
f_test_neg = open(filename+"_test.n", "w")

with open('adult.data', 'r') as file:
    reader = csv.reader(file)
    for _ in range(train_size):
        _ = next(reader)
        
    for _ in range(test_size):
        curr = np.array(next(reader))
        row = curr[indices]
        
        if any(word == ' ?' for word in row): continue

        if curr[-1] == " >50K":
            #print(curr[-1])
            f_test_pos.write(filename+"("+",".join(map(lambda x: _format(x), row))+").\n")
        else:
            f_test_neg.write(filename+"("+",".join(map(lambda x: _format(x), row))+").\n")
f_test_pos.close()
f_test_neg.close()




f_background = open(filename+"_background.txt", "w")
store = {}
with open('adult.data', 'r') as file:
    reader = csv.reader(file)
    #names = next(reader)[:-1] 
      
    for i in range(train_size):
        curr = np.array(next(reader))
        row = curr[indices]
        if any(word == ' ?' for word in row): continue
        for att, val in zip(labels, row):
            try:
                store[att].add(val)
            except KeyError:
                store[att] = {val}
        
                
for key in store.keys():
    #print(key)
    for value in store[key]:
        f_background.write(key.lower()+"("+_format(value)+"). ")
    f_background.write("\n")
        
for l in labels:
    f_background.write(":- modeb(1,(+"+l+" = #"+l+")).\n")
    
f_background.close()
            
```
